What is a dataframe?
 
Dataframe is a 2-D array which has data  ategories in every column.

```python
#Import library
import pandas as pd
```

```python
#Create a serie
 
serie = pd.Series([10,9,8,7,6])
```

```python
#Check values of a serie
 
serie.values
```

```python
#Check index
 
serie.index
```

```python
#Check shape
serie.shape
```

```python
#Create a serie with my own index
 
serie_2 = pd.Series([1,2,3], index = ['a', 'b', 'c'])
serie_2
```

```python
#Call an element
 
#serie_2['a']
serie_2['a':'b']
```

```python
#Create a dictionary
 
dict_data = {'CO': 100, 'MX': 200, 'AR': 300}
```

```python
#Access key's dictionary
 
dict_data.keys()
```

```python
#Extract a value
 
dict_data['MX']
```

```python
#Create a serie from a dictionary
 
pd.Series(dict_data)
```

```python
#Add a non-existent index to a serie
 
serie_3 = pd.Series(dict_data, index=['CO', 'MX', 'PE'])
serie_3
```

```python
#Check NaN or null values of a serie
 
serie_3.isnull()
```
