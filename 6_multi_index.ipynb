# Multi index

```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
#Import data

df_pob = pd.read_csv('population.csv', skiprows=4)
df_pob
```

```python
#Lists columen for deleting some of them.

list(df_pob)
```

```python
#Drop unnecesary columns

df_pob = df_pob.drop(columns=['Country Code',
 'Indicator Name',
 'Indicator Code', 'Unnamed: 65'])
df_pob
```

```python
# Pivot rows to convert them into columns

df = pd.melt(df_pob, id_vars=['Country Name'], var_name='year', value_name='pob')
df = df.sort_values(by=['year','Country Name'])
df
```

```python
#Change number format

pd.options.display.float_format = '{:,.1f}'.format
```

```python
#Change year to categorical value

df['year'] = pd.Categorical(df['year'].apply(str))
df
```

```python
df.dtypes
```

```python
#Filter to do a smaller db

idx_filtro = df['Country Name'].isin(['Aruba', 'Colombia'])
idx_filtro
```

```python
#Create dataframe with the previous filter

df_sample = df[idx_filtro]
df_sample
```

```python
#Create multiple index

df_sample = df_sample.set_index(['Country Name', 'year']).sort_index()
df_sample
```

```python
#Select items just from Colombia

df_sample.loc['Colombia',:]
```

```python
#Select a specific year

df_sample.loc['Colombia', :].loc['2016',:]
```

```python
#Use xs function for the same purpose

df_sample.xs(['Aruba'])
```

```python
#Use xs function to filter by country and year

df_sample.xs(['Aruba', '2018'])
```

```python
#Filter using xs and level

df_sample.xs('2018', level='year')
```

```python
#Apply the multi-index for the whole dataframe

df_countries = df.set_index(['Country Name', 'year']).sort_index()
df_countries
```

```python
#Use slices of de dataframes

ids = pd.IndexSlice
df_countries.loc[ids['Aruba':'Austria', '2015':'2017'], :].sort_index()
```

```python
#Select the indices of level 0

df_countries.index.get_level_values(0)
```

```python
#Select the indices of level 1

df_countries.index.get_level_values(1)
```

```python
#select specific information

df_countries['pop']['Colombia']['2018']
```

```python
#Sum the population of every country in a specific year

df_countries.sum(level = 'year')
```

```python
#Change column to row

df_sample.unstack('year')
```
