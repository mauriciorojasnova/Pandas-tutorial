What is a dataframe?

A dataframe is a 2D array which has categories in each columns.

```python
import pandas as pd
pd.__version__
```

```python
#Create a dict

dict = {'CH': [100, 800, 200], 'CO': [100, 200, 300], 'MX': [300, 500, 400]}
dict
```

```python
#Convert dict into a dataframe

df = pd.DataFrame(dict)
df
```

```python
#create a complex dictionary

import numpy as np

dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
dict_data
```

```python
df = pd.DataFrame(dict_data, index = ['ana','benito','camilo','daniel','erika','fabian','gabriela'])
df
```

```python
#Check dataframe index

df.index
```

```python
#Check column categories

df.columns
```

```python
#Check values

df.values
```

```python
#Select specific columns

df[['edad', 'cm']]
```

```python
#Select a specific row with specific columns

df.loc['ana', ['edad', 'cm']]
```

```python
#Select different rows with specific columns

df.loc[['ana', 'erika'], ['edad', 'cm']]
```

```python
#Select one value knowing row and column

df.loc['daniel', 'Q1']
```

```python
#Select value knowing position

df.iloc[2,1]
```

```python
#Select different values of an index

df.iloc[2, [1, 3]]
```

```python
#Select all indexes or rows

df.iloc[:, [1, 3]]
```

```python
#Select data using conditionals

df[df['edad'] >= 12]
```

```python
#Select data using two conditionals

df[(df['edad'] >= 12) & (df['pais'] == 'mx')]
```

```python
#Another way of selecting data is using queries

df.query('edad>12')
```

```python
#Selecr data using more conditionals

df[df['Q2'] >= df['Q1']]
```
