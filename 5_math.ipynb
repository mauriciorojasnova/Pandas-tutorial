#Math in pandas

```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
df_lmerged = pd.read_csv('london_merged.csv')
df_lmerged.head()
```

```python
#Dataframe variable types

df_lmerged.dtypes
```

```python
#Change timestamp variable to timestamp format

df_lmerged['timestamp'] = pd.to_datetime(df_lmerged['timestamp'])
```

```python
#Change timestamp to hour and create a new column called hour

df_lmerged['hour'] = df_lmerged['timestamp'].dt.hour
df_lmerged['hour']
```

```python
#Delete timestamp column

df = df_lmerged.iloc[:,1:]
df
```

```python
#Change a variable to the square

df['wind_speed']**2
```

```python
#Apply sin function to a column

np.sin(df['wind_speed'])
```

```python
#Subtraction between columns

df['t1'] - df['t2']
```

```python
#Division between columns

df['t1']/df['t2']
```

```python
#Scalar product

df['t1'].dot(df['t2'])
```

```python
#functions

def fun_1(x):
  y = x**2+1
  return y
```

```python
#Apply previous function in a column

df['hour'].apply(fun_1)
```

```python
#Complex function

def fun_2(x, a=10, b=100):
  y = x**2 + a*x + b
  return y
```

```python
#Apply complex function

df['hour'].apply(fun_2, args=[20,200])
```

```python
#Use lambda functions

df['t1'].apply(lambda x: x+273)
```

```python
#lambda function to find mean value

df.apply(lambda x: x.mean())
```

```python
#lambda function to find standard deviation

df.apply(lambda x: x.std())
```

```python
#lambda function to change all values

df.applymap(lambda x: x/100)
```
