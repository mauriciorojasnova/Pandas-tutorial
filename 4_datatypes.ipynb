```python
#Data variety

df_meteorites.nunique()
```

```python
#check nametype and fall variables

df_meteorites[['nametype', 'fall']].astype('category')
```

```python
#Change nametype and fall to category type

df_meteorites[['nametype', 'fall']] = df_meteorites[['nametype', 'fall']].astype('category')
```

```python
#Check new type of variables nametype and fall

df_meteorites.dtypes
```

```python
#Type 'fall'

df_meteorites['fall'].unique()
```

```python
#Count categories in fall

df_meteorites['fall'].value_counts()
```

```python
#Create dummies of the categorical variables

pd.get_dummies(df_meteorites['fall'])
```

```python
#Create new variables in a new column

df_meteorites['ones'] = 1
df_meteorites
```

```python
#Create new variables from a categorical variable

df_meteorites[['fell', 'found']] = pd.get_dummies(df_meteorites['fall'])
df_meteorites
```

```python
#Transform string to time format for the 'year' variable

pd.to_datetime(
    df_meteorites['year'],
    errors = 'coerce',
    format = '%m/%d/%Y %H:%M:%S %p'
)
```

```python
#Overwrite year to the new format

df_meteorites['year'] = pd.to_datetime(
    df_meteorites['year'],
    errors = 'coerce',
    format = '%m/%d/%Y %H:%M:%S %p'
)
df_meteorites.dtypes
```

```python
#Rename a column

df_meteorites.rename(columns={'mass (g)': 'mass'}, inplace=True)
list(df_meteorites)
```

```python
#Delete a column

df_meteorites.drop(['ones'], axis=1, inplace=True)
df_meteorites.head()
```

```python
#Delete rows

df_meteorites.drop([0, 2, 4], axis=0, inplace=True)
df_meteorites.head()
```
