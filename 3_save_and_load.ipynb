##Save data

```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
#Create dict and convert to dataframe

dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}

df = pd.DataFrame(dict_data)
df
```

```python
#Create file for storing data

dir_pandas = '/content/drive/My Drive/Colab Notebooks/pandas_platzi/db/{}'.format('test.csv')
```

```python
#Save df in csv file
df.to_csv(dir_pandas)
```

```python
#Check files in the directory

!ls /content/drive/My\ Drive/Colab\ Notebooks/pandas_platzi/db
```

```python
#Save df in csv file without index
df.to_csv(dir_pandas, index=False)
```

```python
#Save df in excel file

dir_pandas = '/content/drive/My Drive/Colab Notebooks/pandas_platzi/db/{}'
df.to_excel(dir_pandas.format('test.xlsx'))

#Add index=False for removing index column
```

```python
#Save json file

dir_pandas = '/content/drive/My Drive/Colab Notebooks/pandas_platzi/db/{}'
df.to_json(dir_pandas.format('test.json'))

#Add index=False for removing index column
```

```python
#Save pickle file

dir_pandas = '/content/drive/My Drive/Colab Notebooks/pandas_platzi/db/{}'
df.to_pickle(dir_pandas.format('test.pkl'))

#Add index=False for removing index column
```

```python
#Save parquet file

dir_pandas = '/content/drive/My Drive/Colab Notebooks/pandas_platzi/db/{}'
df.to_parquet(dir_pandas.format('test.parquet'))

#Add index=False for removing index column
```
